## Data Cleaning and Preprocessing:

**Strategy:**
1.  Identify and list all ISBNs (across all four tabs of each file) that have sales data recorded after July 1, 2024.
**For dates before July 1, 2024**:
1. Resampling: Convert the weekly data to a consistent frequency (weekly), filling in missing weeks with zero sales. This is crucial for time series analysis.
1. Data Type Conversion: Ensure that 'ISBN' columns are treated as strings and date columns are converted to the datetime type. Set the date column as the index of the DataFrames.

```python
def clean_and_prepare_data_improved(df_isbn_list, df_uk_weekly):
    """
    Cleans and prepares the ISBN list and UK weekly sales data,
    handling missing values more appropriately.

    Args:
        df_isbn_list (pd.DataFrame): DataFrame containing the ISBN list.
        df_uk_weekly (pd.DataFrame): DataFrame containing the UK weekly sales data.

    Returns:
        pd.DataFrame: The cleaned and prepared DataFrame.
    """

    # Convert ISBNs to strings
    df_isbn_list['ISBN'] = df_isbn_list['ISBN'].astype(str)
    df_uk_weekly['ISBN'] = df_uk_weekly['ISBN'].astype(str)

    # Convert dates to datetime objects
    df_isbn_list['Publication Date'] = pd.to_datetime(df_isbn_list['Publication Date'], errors='coerce')
    df_uk_weekly['End Date'] = pd.to_datetime(df_uk_weekly['End Date'], errors='coerce')

    # Merge datasets
    df = pd.merge(df_uk_weekly, df_isbn_list, on='ISBN', how='left', suffixes=('_weekly', '_isbn'))

    # Resample weekly data
    df.set_index('End Date', inplace=True)
    df = df.groupby('ISBN').resample('W').agg({
        'Volume': 'sum',
        'Value': 'sum',
        'Title_weekly': 'first',
        'Author_weekly': 'first',
        'Interval': 'first',
        'ASP': 'mean',
        'RRP_weekly': 'first',  # Use 'first' to preserve the original logic
        'Binding_weekly': 'first',
        'Imprint_weekly': 'first',
        'Publisher Group_weekly': 'first',
        'Product Class_weekly': 'first',
        'Title_isbn': 'first',
        'Author_isbn': 'first',
        'Imprint_isbn': 'first',
        'Publisher Group_isbn': 'first',
        'RRP_isbn': 'first',  # Use 'first' to preserve the original logic
        'Binding_isbn': 'first',
        'Publication Date': 'first',
        'Product Class_isbn': 'first',
        'Country of Publication': 'first'
    })

    # Reset index
    df.reset_index(inplace=True)

    # Forward fill and backward fill for specific columns within each ISBN group
    # Only apply to columns that exist in df
    existing_columns = set(df.columns)
    columns_to_fill = ['RRP_weekly', 'Title_weekly', 'Author_weekly', 'Binding_weekly', 'Imprint_weekly',
                       'Publisher Group_weekly', 'Product Class_weekly', 'RRP_isbn']
    columns_to_fill = [col for col in columns_to_fill if col in existing_columns]

    for col in columns_to_fill:
        df[col] = df.groupby('ISBN')[col].ffill().bfill()

    # Filter out rows with ISBN '0'
    df = df[df['ISBN'] != '0']

    # Create df_filtered containing only data up to '2024-07-01'
    df_filtered = df[df['End Date'] <= '2024-07-01'].copy()

    # --- Impute remaining missing values ---
    # Impute remaining missing values in categorical columns using the most frequent value
    cat_cols = ['Author_weekly', 'Title_weekly', 'Binding_weekly', 'Imprint_weekly',
                'Publisher Group_weekly', 'Product Class_weekly']

    # Only impute columns that exist in df_filtered
    cat_cols = [col for col in cat_cols if col in df_filtered.columns]

    for col in cat_cols:
        most_frequent_value = df_filtered[col].mode()[0]
        df_filtered[col].fillna(most_frequent_value, inplace=True)

    # Impute remaining missing values in 'RRP_weekly' with the median value
    if 'RRP_weekly' in df_filtered.columns:
        median_rrp = df_filtered['RRP_weekly'].median()
        df_filtered['RRP_weekly'].fillna(median_rrp, inplace=True)

    # --- Feature Engineering ---

    # Convert 'Interval' to integer after handling missing values
    df_filtered['Interval'] = df_filtered['Interval'].fillna(0).astype(int)

    # Extract Year and Week from Interval
    df_filtered['Year'] = df_filtered['Interval'].apply(lambda x: int(str(x)[:4]) if x > 0 else 0)
    df_filtered['Week'] = df_filtered['Interval'].apply(lambda x: int(str(x)[4:]) if x > 0 else 0)

    # --- Handle Numerical Columns ---

    # Zero out negative values in df_filtered for specific columns
    num_cols = ['Volume', 'Value', 'ASP']
    df_filtered.loc[:, num_cols] = df_filtered[num_cols].clip(lower=0)

    return df_filtered

# Clean and prepare the data
df_filtered = clean_and_prepare_data_improved(df_isbn_list, df_uk_weekly)
```
