```python
# Function to load and combine data from a single file
def load_and_combine_data(file_path):
    xls = pd.ExcelFile(file_path)
    dfs = []
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name)
        dfs.append(df)
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df

# Load both files

isbn_list_file = '/content/ISBN List.xlsx'
uk_weekly_file = '/content/UK Weekly Trended Timeline from 200101_202429.xlsx'

df_isbn_list = load_and_combine_data(isbn_list_file)
df_uk_weekly = load_and_combine_data(uk_weekly_file)
```

## Data Cleaning and Preprocessing:

```python
# Convert ISBNs to strings
df_isbn_list['ISBN'] = df_isbn_list['ISBN'].astype(str)
df_uk_weekly['ISBN'] = df_uk_weekly['ISBN'].astype(str)

# Convert dates to datetime objects
df_isbn_list['Publication Date'] = pd.to_datetime(df_isbn_list['Publication Date'])
df_uk_weekly['End Date'] = pd.to_datetime(df_uk_weekly['End Date'])

# Merge datasets - keep df_uk_weekly and df_isbn_list separate
df = pd.merge(df_uk_weekly, df_isbn_list, on='ISBN', how='left', suffixes=('_weekly', '_isbn'))

# Resample weekly data and fill missing weeks with 0 - do this on df
df.set_index('End Date', inplace=True)
df = df.groupby('ISBN').resample('W').agg({
    'Volume': 'sum',
    'Value': 'sum',
    'Title_weekly': 'first',
    'Author_weekly': 'first',
    'Interval': 'first',
    'ASP': 'mean',
    'RRP_weekly': 'mean',
    'Binding_weekly': 'first',
    'Imprint_weekly': 'first',
    'Publisher Group_weekly': 'first',
    'Product Class_weekly': 'first',
    'Title_isbn': 'first',
    'Author_isbn': 'first',
    'Imprint_isbn': 'first',
    'Publisher Group_isbn': 'first',
    'RRP_isbn': 'mean',
    'Binding_isbn': 'first',
    'Publication Date': 'first',
    'Product Class_isbn': 'first',
    'Country of Publication': 'first'
}).fillna(0)

df.reset_index(inplace=True)

# Filter out rows with ISBN '0'
df = df[df['ISBN'] != '0']

# Impute missing 'Author_weekly' using 'Author_isbn'
df['Author_weekly'] = df['Author_weekly'].replace('', np.nan) # replace blanks with NAN
df['Author_weekly'] = df['Author_weekly'].fillna(df['Author_isbn'])

# Impute missing 'RRP_weekly' using 'RRP_isbn'
df['RRP_weekly'] = df['RRP_weekly'].replace('', np.nan) # replace blanks with NAN
df['RRP_weekly'] = df['RRP_weekly'].fillna(df['RRP_isbn'])

# Filter out ISBNs with sales data beyond 2024-07-01
max_dates = df.groupby('ISBN')['End Date'].max()
isbns_beyond_2024 = max_dates[max_dates > '2024-07-01'].index
print("ISBNs with sales data beyond 2024-07-01:", isbns_beyond_2024)

# Create df_filtered containing only data up to '2024-07-01'
df_filtered = df[df['End Date'] <= '2024-07-01']

# Zero out negative values in df_filtered
df_filtered.loc[df_filtered['Volume'] < 0, 'Volume'] = 0
df_filtered.loc[df_filtered['Value'] < 0, 'Value'] = 0
df_filtered.loc[df_filtered['ASP'] < 0, 'ASP'] = 0
```

## Create a Function to Isolate Outlier Data

```python
def get_outlier_data(df, volume_range=None, value_range=None, asp_range=None):
    """
    Filters the DataFrame to return rows that fall within the specified ranges for Volume, Value, and ASP.

    Args:
        df: The DataFrame to filter (df_negatives).
        volume_range: A tuple (min, max) for the Volume range.
        value_range: A tuple (min, max) for the Value range.
        asp_range: A tuple (min, max) for the ASP range.

    Returns:
        A DataFrame containing only the rows that meet the outlier criteria.
    """
    df_outliers = df.copy()  # Create a copy to avoid modifying the original

    if volume_range:
        df_outliers = df_outliers[(df_outliers['Volume'] >= volume_range[0]) & (df_outliers['Volume'] <= volume_range[1])]
    if value_range:
        df_outliers = df_outliers[(df_outliers['Value'] >= value_range[0]) & (df_outliers['Value'] <= value_range[1])]
    if asp_range:
        df_outliers = df_outliers[(df_outliers['ASP'] >= asp_range[0]) & (df_outliers['ASP'] <= asp_range[1])]

    return df_outliers
```
